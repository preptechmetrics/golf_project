**Modify and Analyze Data**

```python
# Aggregate holes #1 - #18 and find the mean of all holes
all_holes = Modified_raw_data_df.columns.str.contains('Hole')
Modified_raw_data_df.loc[:, all_holes].mean(axis=1)
```

```python
# Sum the scores submitted for holes #1 - 18 and call it 'Calc_score' (league is 9 holes per league date)
Modified_raw_data_df['Calc_score']= Modified_raw_data_df.iloc[:, 4:-2].sum(axis=1)
Modified_raw_data_df
```

```python
# Compare the total submitted score with the calculated score from the sum of each hole submission
compare_scores = np.where(Modified_raw_data_df["Submitted_Score"] == Modified_raw_data_df["Calc_score"], True, False)
print(compare_scores)
```

```python
# Display the comparison from above by date and player
Modified_raw_data_df["Equal"] = compare_scores
print(Modified_raw_data_df[['PlayDate', 'Golfer1', 'Submitted_Score', 'Calc_score', 'Equal']])
```

```python
# Identify Submitted Scores and Calculated scores that didn't match 
Calc_error = Modified_raw_data_df[Modified_raw_data_df['Equal']== False]
print(Calc_error[['PlayDate', 'Golfer1', 'Submitted_Score', 'Calc_score', 'Equal']])
```

```python
# Use .describe to get descriptive statistics
summary_data = Modified_raw_data_df.describe()
summary_data
```

```python
# Take a closer look at minimums from the summary data to determine if there were any "hole-in-one"'s
summary_data['Min_score']= summary_data.min(axis=1)
print(summary_data['Min_score'])
```
